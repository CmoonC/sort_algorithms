```python
def bubble(A):
    n = len(A)
    while True:
        swapped = False
        for i in range(1, n):
            if A[i - 1] > A[i]:
                A[i - 1], A[i] = A[i], A[i - 1]
                swapped = True
        if not swapped: break    
```

```python
def selection(A):
    n = len(A)
    for i in range(n - 1):
        for j in range(i + 1, n):
            if A[j] < A[i]:
                A[j], A[i] = A[i], A[j]
```

```python
def quick(A, p, q):
    if p < q:
        i = part(A, p, q)
        quick(A, p, i - 1)
        quick(A, i + 1, q)

def part(A, p, q):
    x = A[q]
    i = p - 1
    for j in range(p, q):
        if A[j] < x:
            i += 1
            A[i], A[j] = A[j], A[i]
    A[i + 1], A[q] = A[q], A[i + 1]
    return i + 1
```

```python
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
n=np.random.normal(23,9,1000)
```

```python
results1=[]
results2=[]
results3=[]
for i in range(100):
    np.random.shuffle(n)
    start1=time.time()
    bubble(n)
    result1=time.time()-start1
    results1.append(result1)
    result1 =0
    start1 =0
    
```

```python
for i in range(100):
    np.random.shuffle(n)
    start2=time.time()
    selection(n)
    result2=time.time()-start2
    results2.append(result2)
    result2 =0
    start2 =0
 
```

```python
for i in range(100):
    np.random.shuffle(n)
    start3=time.time()
    quick(n,0,999)
    result3=time.time()-start3
    results3.append(result3)
    result3 =0
    start3 =0
 
```
